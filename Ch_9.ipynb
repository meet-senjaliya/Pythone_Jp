## Inheritance

- A class can inherit from other class
- It improves code reuse
- Constructor, attributes and methods are inherited to child class
- Parent class has no access to the child class
- child class methods overrides the parent class methods
    - this is called method overriding
-```super()``` method is used to invoke parent class method and constructor

## `super() method`

- If we write constructor in both parent and child class, the parent class is not available in child class 
- `super()` methods is used to call parent class from the child

`````python
class phone:
    def __init__ (self,price,brand,camera):
        print("inside phone class")
        self.price = price
        self.brand = brand
        self.camera = camera
    def buy(self):
        print("Buying phone")
class smartphone(phone):
    def buy(self):
        print("Buying smartphone")
        super().buy()
`````

`````python
s = smartphone(20000,"iphone",13)
`````

`````python
s.buy()
`````

## METHOD RESOLUTION OREDER IN PYHTON
- It defines an order in which the class is searched in case of multiple inheritance.
- It is based on C3 Linearization Algorithm(>= Python 2.3)

## C3 Linearization Algorithm
**Notation**

- For a list of classes-C1C2C3...Cn
    - the C1 is the head of class and the rest is tall
- C3 Linearization Algorithm is defined as MRO of sum of any class C and the linear of Parents and the list of parents
```
    L(C(B1,B2...Bn)) = C + merge(L(B1),L(B2),...Bn))
```
- Here B1 to Bn are the parent classes of C

1. Take the head of list L(B1) from left to right
2. If the head of selected class is not in tail of the list , add it to the linear and remove it from the list.
3. Repeat the process untill the classes are removed or it impossible to find good head.

`````python

class F:
    pass
class E:
    pass
class D:
    pass
class C(D,F):
    pass
class B(D,E):
    pass
class A(B,C):
    pass
`````

`````python
A.mro()
`````

## First Step 
- MRO OF D,E,F,O
- L(D) = DO
- L(E) = EO
- L(F) = FO
- L(O) = O

## Second Step
```
L(C) = C+ merge(L(D),L(F),DF)
     = C + merge(DO,FO,DF)
     = C + D + merge(O,FO,F)
     = C + D + F + O
     = CDFO
```

## Third Step
```
L(B) = B + merge(L(D),L(E),DE)
     = B + MERGE(DO, EO, DE)
     = B + D + MERGE(O, EO, E)
     = B + D + E + O
     = BDEO
````

## Four Step
```
L(A) = A + MERGE(L(B),L(C),BC)
     = A + MERGE(BDEO,CDFO,BC)
     = A + B + MERGE(DEO , CDFO , C)
     = A + B + C + MERGE(DEO, DFO)
     = A + B+ C + D + MERGE(EO, FO)
     = A + B+ C + D + E + F + O
```

## OPERATOR OVERLOADING
```
__add__ = +
__sub__ = -
__mul__ = *
__truediv__ = /
```

`````python
class Fraction:
    
    def __init__(self,x,y):
        self.num = x
        self.den = y
    
    def __add__(self,other):
        new_num = self.num*other.den + other.num*self.den
        new_den = self.den * other.den
        return f"{new_num}/{new_den}"
    
    def __sub__(self,other):
        new_num = self.num*other.den - other.num*self.den
        new_den = self.den * other.den
        return f"{new_num}/{new_den}"
    
    def __mul__(self,other):
        new_num = self.num*other.num 
        new_den = self.den * other.den
        return f"{new_num}/{new_den}"
    
    def __truediv__(self,other):
        new_num = self.num*other.den 
        new_den = self.den * other.num
        return f"{new_num}/{new_den}"
    
        
        
`````
