```python
import tensorflow
from tensorflow import keras
```

```python
(x_train,y_train), (x_test,y_test) = keras.datasets.mnist.load_data()
```

```python
x_train.shape #60k rows and 28*28 pixeal na 2d araay na image
```

```python
x_train[1]
```

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[2])
plt.show()
```

```python
y_train
```

```python
x_train = x_train/255
x_test= x_test/255
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D

model = Sequential()
model.add(Conv2D(64, (3,3), activation='relu', input_shape=(28,28,1)))
model.add(MaxPooling2D((2,2)))
model.add(Conv2D(32, (3,3), activation='relu'))
model.add(MaxPooling2D((2,2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(10, activation='softmax'))

```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='sgd',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(x_train, y_train, epochs=10, validation_split=0.2)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred.argmax(axis=1))
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.show()

```

```python
import pandas as pd

train_df = pd.read_csv('sign_mnist_train.csv')
test_df = pd.read_csv('sign_mnist_valid.csv')
```

```python
y_train = train_df['label']
y_test = test_df['label']

x_train = train_df.drop('label', axis=1)
x_test = test_df.drop('label', axis=1)
```

```python
y_train
```

```python
y_train = keras.utils.to_categorical(y_train, num_classes=24)
y_test = keras.utils.to_categorical(y_test, num_classes=24)
```
