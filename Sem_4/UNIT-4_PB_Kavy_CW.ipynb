### 218

Predict salary based on job position of 6.5 using polynomial regression with a degree of 3 and scikit learn library for the given 'Position_Salaries.csv' dataset. (Don't split data for training/testing)

---

```python
# import libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# import dataset
df = pd.read_csv("https://raw.githubusercontent.com/patelmanishv/Sem4Data/refs/heads/main/Datasets/Position_Salaries.csv")
df.head(2)

# Define features and target
x = df[["Level"]]
y = df["Salary"]
print(x.shape)
print(y.shape)

# Create polynomial feature
poly = PolynomialFeatures(degree=3)
x_poly = poly.fit_transform(x)

# Create and fit model
model = LinearRegression()
model.fit(x_poly,y)

# Prediction
level = np.array([[6.5]])
level_poly = poly.fit_transform(level)
predicted_salary = model.predict(level_poly)
print(predicted_salary)
```

### 219

For x = np.arange(0, 30) and y = np.array([3, 4, 5, 7, 10, 8, 9, 10, 10, 23, 27, 44, 50, 63, 67, 60, 62, 70, 75, 88, 81, 87, 95, 100, 108, 135, 151, 160, 169, 179]), apply polynomial regression using scikit learn library and calculate R squared, coeficient and intercept. Predict the y values for x = np.arange(5). (Don't split data for training/testing)

---

```python
# import libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Define features and target
x = np.arange(0, 30).reshape(-1,1)
y = np.array([3, 4, 5, 7, 10, 8, 9, 10, 10, 23, 27, 44, 50, 63, 67, 60, 62, 70, 75, 88, 81, 87, 95, 100, 108, 135, 151, 160, 169, 179])
print(x.shape)
print(y.shape)

# Create polynomial feature
poly = PolynomialFeatures(degree=3) # by default degree=3
x_poly = poly.fit_transform(x)

# Create and fit model
model = LinearRegression()
model.fit(x_poly,y)

# Evaluation metrics
print("R2 score =",model.score(x_poly,y))
print("The coefficients =",model.coef_)
print("The intercept =",model.intercept_)

# Prediction
level = np.arange(5).reshape(-1,1)
level_poly = poly.fit_transform(level)
predicted_values = model.predict(level_poly)
print(predicted_values)
```

```python
# 2. Load RealEstate dataset, create a dataframe and check datatypes of its attributes using appropriate method.
df = pd.read_csv("https://raw.githubusercontent.com/patelmanishv/Sem4Data/refs/heads/main/Datasets/RealEstate.csv")
print(df.info())
```

### 230

Write a python program for following tasks -

Make sure all the required libraries are imported at the beginning of the Jupyter Notebook.

a) Load the windpower dataset into the notebook.
b) Check records for any missing values and remove them permanently using appropriate method.
c) Treat windspeed as the independent variable and power as the dependent variable. Split the dataset into training (75%) and testing (25%) sets using random state of 42.
d) Use scikit-learn to train polynomial regression models of degrees 3, 4, 5, and 6.
e) Predict power values on the test set and print the Mean Squared Error (MSE) and R² score (coefficient of determination) for each polynomial degree.
f) Plot actual vs. predicted power values for all polynomial degrees on the same graph for comparison and label the plot accordingly.
g) Create a plot showing MSE vs. polynomial degree. Which degree has the lowest MSE? Write in the comment.
h) Print the coefficients and intercept for each trained polynomial regression model.
i) Use each model to predict power output for wind speeds of 5, 10, 15, and 20 m/s, and print the predictions.

---

```python
# a) Load the windpower dataset into the notebook.
df = pd.read_csv("https://raw.githubusercontent.com/patelmanishv/Sem4Data/refs/heads/main/Datasets/windpower.csv")
df.head(2)

# b) Check records for any missing values and remove them permanently using appropriate method.
df.isnull().sum()

# Remove the NA values permanently
df.dropna(inplace=True)

df.isnull().sum()

print(df.to_string())

#  c) Treat windspeed as the independent variable and power as the dependent variable. Split the dataset into training (75%) and testing (25%) sets using random state of 42.
x = df[["Wind speed (m/s)"]]
y = df["Power (kW)"]
print(x.shape)
print(y.shape)

# Split the dataset into training (75%) and testing (25%) sets using random state of 42.
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)

# d) Use scikit-learn to train polynomial regression models of degrees 3, 4, 5, and 6.
from sklearn.metrics import r2_score,mean_squared_error
import seaborn as sns
import matplotlib.pyplot as plt
givedegrees = [3,4,5,6]
mse_values = []
r2_scores = []

windspeeds = np.array([[5],[10],[15],[20]])
windprediction = {}

# plot
plt.style.use('ggplot') # optional
plt.figure(figsize=(12,8)) # optional
plt.title("Polynomial Regression of Windpower")
plt.scatter(x,y,color='k',label='truevalues')

for degree in givedegrees:
    poly = PolynomialFeatures(degree=degree)
    x_train_poly = poly.fit_transform(x_train)
    x_test_poly = poly.fit_transform(x_test) 
    model = LinearRegression()
    model.fit(x_train_poly,y_train)
    predictions = model.predict(x_test_poly)
    mse = mean_squared_error(y_test,predictions)
    r2 = r2_score(y_test,predictions)
    mse_values.append(mse)
    r2_scores.append(r2)
    print(f"Result for {degree} polynomial regression")
    print(f"Coefficients: {model.coef_}")
    print(f"Intercepts: {model.intercept_}")
    print(f"Mean Squared Error: {mse:.2f}")
    print(f"Coefficient of Determination: {r2:.2f}")

    plt.plot(x_test,predictions,'x',label=f'predictedvalues for {degree}')
    windspeedpoly = poly.fit_transform(windspeeds)
    powerpredict = model.predict(windspeedpoly)
    windprediction[degree] = powerpredict
plt.xlabel("Windspeed")
plt.ylabel("Power")
plt.legend()
plt.show()

print(windprediction)

# g) Create a plot showing MSE vs. polynomial degree. Which degree has the lowest MSE? Write in the comment.
plt.figure(figsize=(8,6))
plt.plot(givedegrees,mse_values,marker='o',linestyle='-',color='b')
plt.title("MSE vs Degree")
plt.xlabel("Degree")
plt.ylabel("MSE")
plt.grid(True)
plt.show()
# The sixthdegree has the lowest MSE of all
```
