## Unit-3 Feature Creation & Feature Transformation 

## To Convert Categorical Columns to Numerical Columns 

```python
import pandas as pd
```

```python
data = {
    'Marks':[40,42,42,41,43,45,44,49,45,42,46,48],
    'Gender':['Male','Female','Male','Female','Male','Female',
              'Male','Female','Male','Female','Male','Female']
}

df = pd.DataFrame(data)
df
```

```python
pd.get_dummies(df) #One-hot Encoding
```

```python
pd.get_dummies(df, drop_first = True) #One-hot Encoding
```

## Feature Creation

```python
df
```

```python
import numpy as np
df['Result'] = np.where(df['Marks']>45,'Pass','Fail')
```

```python
df
```

```python

df['Grade'] = np.where(df['Marks'] > 47, 'A',        # if
               np.where(df['Marks'] > 45, 'B',       # elif
               np.where(df['Marks'] > 43, 'C', 'D'))) # else

print(df)
```

# Train-Validation Split

```python
from sklearn.model_selection import train_test_split
```

```python
data = {
    'Marks':[40,42,42,41,43,45,44,49,45,42,46,48],
    'Gender':['Male','Female','Male','Female','Male','Female',
              'Male','Female','Male','Female','Male','Female']
}

df = pd.DataFrame(data)
x = df["Gender"]
y = df['Marks']

```

```python
x_train,x_test,y_train,y_test  = train_test_split(x,y,test_size = 0.2, random_state  = 10 ) # rs limit (0-10)
x_train
```
