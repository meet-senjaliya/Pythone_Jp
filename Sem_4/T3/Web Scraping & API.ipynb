```python
response = requests.get('https://subslikescript.com/movies')
content = response.text
data = BeautifulSoup(content,'html.parser')
print(data.prettify())
```

```python
movie_list = data.find('ul',class_="scripts-list")
print(movie_list)
```

```python
l = movie_list.find_all('a')
print(len(l))
```

```python
links = []
for i in l:
    x = i['href']
    links.append(x)
print(links)
```

```python
for i in links:
    url = 'https://subslikescript.com' + i
    response = requests.get(url)
    content = response.text
    data = BeautifulSoup(content,'html.parser')
    title = data.find('h1').text
    script = data.find('div',class_='full-script').text
    f = open(title+'.txt','w',encoding='utf-8')
    f.write(script)
    print(f'{title} created successfully')
    f.close()
```

```python
user = int(input('Enter No. of Pages to scrape: '))
for i in range(1,user+1):
    url = 'https://subslikescript.com/movies?page=' + str(i)
    response = requests.get(url)
    content = response.text
    data = BeautifulSoup(content,'html.parser')
    movie_list = data.find('ul',class_="scripts-list")
    l = movie_list.find_all('a')
    links = []
    for i in l:
        x = i['href']
        links.append(x)
    for i in links:
        url = 'https://subslikescript.com' + i
        response = requests.get(url)
        content = response.text
        data = BeautifulSoup(content,'html.parser')
        title = data.find('h1').text
        script = data.find('div',class_='full-script').text
        f = open(title+'.txt','w',encoding='utf-8')
        f.write(script)
        print(f'{title} created successfully')
        f.close()
```

```python
f = open('newflipkart.html',encoding='utf-8')
data = BeautifulSoup(f, 'html.parser')
print(data.prettify())
```

```python
phones = data.find_all('div',class_="tUxRFH")
print(len(phones))
```

```python
d = {'Name':[],'Color':[],'Memory':[],'Rating':[],'Reviews':[],'AP':[],'DP':[],'Disc':[]}
for i in phones:
    x = i.find('div',class_="KzDlHZ").text
    d['Name'].append(x.split('(')[0].strip())
    d['Color'].append(x.split('(')[1].split(',')[0].strip())
    d['Memory'].append(x.split(',')[-1].strip(')'))
    r = i.find('div',class_="XQDdHH")
    if r != None:
        d['Rating'].append(r.text)
    else:
        d['Rating'].append('')
    rev = i.find('span',class_="Wphh3N").text
    d['Reviews'].append(rev.split()[-2])
    ap = i.find('div',class_="yRaY8j ZYYwLA").text
    ap = ''.join(ap.strip('₹').split(','))
    d['AP'].append(int(ap))
    dp = i.find('div',class_="Nx9bqj _4b5DiR").text
    dp = ''.join(dp.strip('₹').split(','))
    d['DP'].append(int(dp))
    disc = i.find('div',class_="UkUFwK").text.split()[0]
    d['Disc'].append(disc)
print(d)
```

### Pagination

```python
url = 'https://jsonplaceholder.typicode.com/posts'
params = {
    '_page':1,
    '_limit':5
}
response = requests.get(url, params = params)
print(response.json())
```

```python
page = 1
while page<=3:
    params = {
        '_page':page,
        '_limit':5
    }
    response = requests.get(url, params = params)
    print(f'Page: {page}')
    print(response.json())
    page += 1
```

```python
from bs4 import BeautifulSoup
f = open('flipkart (1).html',encoding='utf-8')
data = BeautifulSoup(f,'html.parser')
print(data.prettify())
```

```python
phones = data.find_all('div',class_='tUxRFH')
print(len(phones))
```

```python
d = {'Name':[], 'Memory':[], 'Color':[],'Stars':[],'Reviews':[],
    'Actual Price':[], 'Final Price':[],'Exchange Offer':[], 'Discount':[]}
import numpy as np
for i in phones:
    title = i.find('div',class_='KzDlHZ').text
    index = title.rfind('(')
    d['Name'].append(title[ :index].strip())
    c = title.split('(')[-1].split(',')[0]
    d['Color'].append(c)
    m = title.split('(')[-1].split(',')[1].strip(' )')
    d['Memory'].append(m)
    s = i.find('div',class_='XQDdHH').text
    d['Stars'].append(s)
    r = i.find('span',class_="Wphh3N").text
    d['Reviews'].append(r.split()[-2])
    try:
        ap = i.find('div',class_="yRaY8j ZYYwLA").text
        d['Actual Price'].append(ap.strip('₹'))
    except:
        d['Actual Price'].append(np.nan)
    dp = i.find('div',class_="Nx9bqj _4b5DiR").text
    d['Final Price'].append(dp.strip('₹'))
    try:
        exc = i.find('div',{'style':"color:rgb(0,0,0);font-size:14px;font-style:normal;font-weight:700",'class':"yiggsN O5Fpg8"}).text
        d['Exchange Offer'].append(exc.strip('₹'))
    except:
        d['Exchange Offer'].append(np.nan)
print(d)
```
