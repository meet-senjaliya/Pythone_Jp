## to encode the number in the following formate 

- input : 1034895
- output : 2145906

def encoded(number):
    encodenumber = 0
    multiplier = 1
    
    while number > 0:
        digit = number % 10
        if digit == 9:
            encodenumber += 0*multiplier
        else:
            encodenumber += (digit+1)*multiplier
            
        multiplier *= 10
        number //= 10
    return encodenumber
