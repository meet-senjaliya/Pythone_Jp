````python
# Task 12
df = df.iloc[10:]
````

````python
# Task 13 - Detect outliers
q1 = df["popularity"].quantile(0.25)
q3 = df['popularity'].quantile(0.75)
iqr = q3 - q1
lower = q1 - 1.5*iqr
upper = q3 + 1.5*iqr
outliers = df[(df["popularity"] < lower) | (df["popularity"] > upper)]
print(outliers)
````

````python
# Task 13 - Remove outliers
q1 = df["energy"].quantile(0.25)
q3 = df['energy'].quantile(0.75)
iqr = q3 - q1
lower = q1 - 1.5*iqr
upper = q3 + 1.5*iqr
cleandf = df[(df["energy"] >= lower) & (df["energy"] <= upper)]
print(cleandf)
````

````python
# Task 2 How many Petrol cars are of the year 2017 and whose selling price > 10 lakhs.
len(df[(df["Fuel_Type"] == "Petrol") & 
    (df["Year"] == 2017) & 
    (df["Selling_Price"] >10)])
````

````python
# Task 3.How many swift cars are there with selling price < 4 Lakhs
len(df[(df["Car_Name"] == 'swift') & (df["Selling_Price"] <4)])
````

````python
# Task 4.How many Automatic Transmission Petrol Car are of the year 2015 whose selling price is > 10 Lakhs.
len(df[(df["Transmission"] == "Automatic") & 
(df["Fuel_Type"] == "Petrol") &
(df["Year"] == 2015) &
(df["Selling_Price"] >10)])
````

````python
# 5.List out Vehicles with Automa c Transmission and selling price < 1 Lakh
df[(df["Transmission"] == "Automatic") & 
(df["Selling_Price"] <1)]
````

````python
# Task 6 How Many Petrol Vehicles are there with kms driven more than 50000kms and Year is between 2010 to 2015(both Year included).
len(df[(df["Fuel_Type"] == "Petrol") &
(df["Kms_Driven"] > 50000) &
(df["Year"] <= 2015) &
(df["Year"] >= 2010)  
])
````
