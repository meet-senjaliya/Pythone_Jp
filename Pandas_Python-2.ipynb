## Merge Function

```python
df1 = pd.DataFrame({
    'name':['A','B','C','D'],
    'salary':[25000,30000,22000,35000],
    'dept_id':[10,20,10,30]
})
df2 = pd.DataFrame({
    'dept_id':[10,20,40],
    'dept':['HR','IT','Fin']
})
print(df1)
print(df2)
```

```python
pd.merge(df1,df2,how='inner',on='dept_id')
```

```python
pd.merge(df1,df2,how='outer',on='dept_id')
```

```python
pd.merge(df1,df2,how='left',on='dept_id')
```

```python
pd.merge(df1,df2,how='right',on='dept_id')
```

```python
pd.merge(df1,df2,how='cross')
```

```python
df1 = pd.DataFrame({
    'name':['A','B','C','D'],
    'salary':[25000,30000,22000,35000],
    'dept_id':[10,20,10,30]
})
df2 = pd.DataFrame({
    'id':[10,20,40],
    'dept':['HR','IT','Fin']
})
print(df1)
print(df2)
```

```python
pd.merge(df1,df2,left_on='dept_id',right_on='id',how='outer')
```

```python
pd.merge(df1,df2,left_index=True,right_index=True,how='outer')
```

```python
df1 = pd.DataFrame({
    'name':['A','B','C','D'],
    'salary':[25000,30000,22000,35000],
    'dept_id':[10,20,10,30]
})
df2 = pd.DataFrame({
    'id':[10,20,40],
    'name':['HR','IT','Fin']
})
print(df1)
print(df2)
```

```python
pd.merge(df1,df2,left_index=True,right_index=True,how='outer',suffixes=['_emp','_dept'])
```
