```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Seaborn

```python
df = pd.DataFrame({
    "Student":['A','B','C','D','E'],
    "Math":[78,85,90,95,88],
    "Science":[80,99,70,85,92],
    "English":[80,90,93,115,98],
    "Computer":[75,88,77,78,98],
    "Gender":['M','F','M','F','M'],
    "Studytype":['self','coaching','self','coaching','self'],
    "Hours": [4,5,2,3,1]
})
```

### Heatmap of correlation

```python
corr = df[["Math","Science","English","Computer","Hours"]].corr()
corr
```

```python
sns.heatmap(
    corr,
    vmin = 0.3,
    vmax = 1.0, # color intensity
    cmap = 'viridis', # color schemes
    linewidths=2,
    linecolor="maroon",
    annot=True,
    cbar = True
    )
plt.title("Heatmap of correlation")
plt.show()
```

### BOXPLOT
- It is used to visualise max,min,median and outliers

```python
sns.boxplot(
    x = "Studytype", # Categorical variable
    y = "Math", # Numerical Variable
    data = df,
    hue="Gender"
)
plt.title("Boxplot")
plt.show()
```

### Scatter plot
- It shows the relationship between two numerical variables

```python
sns.scatterplot(
    x = "Hours",
    y = "Math",
    data = df,
    hue="Gender",
    style="Studytype",
    s = 100
)
plt.title("Scatter plot")
plt.show()
```

## PLOTLY LIBRARY
- To create interactive visualization in python

### import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.DataFrame({
    "Category":['A','B','C','D','E'],
    "Sales":[100,200,150,250,120],
    "Profit":[20,-30,40,-10,35],
    "Region":['east','west','east','west','east'],
    "Gender":['M','F','M','F','M'],
    "Type":['self','coaching','self','coaching','self'],
    "Hours":[4,3,5,2,1],
    "Marks":[50,55,67,75,44],
    "Attempts":[2,1,3,2,1]
})
```

## Simple bar plot

```python
px.bar(
    df,
    x = "Category",
    y = "Sales",
    width=500, # fix size
    height=400,
    title = "Simple bar plot", # title of plot
    color = "Region", # classify based on region
    hover_data=["Profit","Marks"], # extra hover detail
    pattern_shape="Region", # add pattern
    barmode = 'group', #group bar
    text_auto=True, #add text to bar
    text = "Attempts", # text data
    orientation='v', # v or h orientation
    color_discrete_map={"east":"red","west":"green"}, # color control
    color_discrete_sequence=["red","green"] 
).show()
```

```python
### Grouped bar plot
px.bar(
df,
x = "Category",
y = ["Sales","Profit"],
barmode = "group",
title = "Grouped Bar chart"
).show()
```

```python
## Diverging bar
px.bar(
    df,
    x = "Category",
    y = "Profit",
    color = "Profit",
    barmode = "relative",
    title = "A diverging barplot"
).show()
```

## SCATTER PLOT
- It shows visually the relationship between two numerical variables

```python
px.scatter(
    df,
    x = "Hours", # Numerical Variable
    y = "Marks", # Numerical variable
    title = "A Scatter Plot",
    color = "Gender",
    size = "Attempts",
    symbol = "Type",
    facet_col="Region"
).show()
```

### 3D scatter

```python
px.scatter_3d(
    df,
    x = "Hours",
    y = "Marks",
    z = "Attempts",
    color = "Region",
    size = "Attempts",
    title = "3D scatter plot"
).show()
```

### SCATTER MATRIX

```python
px.scatter_matrix(
    df,
    dimensions = ["Hours","Marks","Attempts","Sales"],
    title = "Scatter Matrix",
    color = "Region",
).show()
```

## LINE CHART

```python
px.line(
    df,
    x = "Hours",
    y = ["Sales","Profit"],
    color = "Region",
    markers = True,
    hover_data=["Profit", "Marks"],
    title = "a line chart"
).show()
```

## PIE CHART

```python
fig = px.pie(
    df,
    values = "Sales",
    names = "Category",
    title = "Pie Chart"
)
```

```python
fig.update_traces(
    textinfo = "percent+label",
    textposition = 'inside',
    pull = [0,0.1,0,0,0]
)
fig.show()
```

### HISTOGRAM

```python
px.histogram(
    df,
    x = "Marks",
    y = None,
    nbins = 5,
    color = "Gender",
    histfunc="count",
    pattern_shape="Region",
    text_auto=True,
    ).show()
```

## NETWORKX
- This library is used to create and analyse graphs in Graph Theory

```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Create undirected graph object
G = nx.DiGraph()
# Add nodes
G.add_node(1)
G.add_nodes_from([2,3])
G.add_nodes_from(range(4,7))
# Add edges
G.add_edge(1,2)
G.add_edge(1,1)
G.add_edges_from([(2,3),(3,6),(4,6),(5,6)])
nx.draw(G,with_labels=True)
plt.show()
```

```python
g = nx.DiGraph()
g.add_nodes_from(['A','B','C','D'])
g.add_edges_from([('A','B'),('B','C'),('B','D'),('D','A'),('C','D')])
nx.draw_networkx(g,with_labels = True,
       node_size = 1000,node_color='red',
       edge_color = "green")
plt.show()
```

### GRAPH BASED ON ADJACENT MATRIX

```python
g = nx.DiGraph()
L = [[1,0,0,1,0,1,1],
     [0,1,1,1,1,1,0],
     [1,1,1,0,1,0,1]]
g.add_nodes_from([0,1,2])
for i in range(0,3):
    for j in range(0,3):
        if L[i][j] == 1:
            g.add_edge(i,j)
nx.draw(g,with_labels=True)
plt.show()
```
