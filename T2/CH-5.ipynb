## 

````python
def longest(nums):
    result = 0
    nums = set(nums)
    for num in nums:
        if num-1 not in nums:
            current = num
            length = 1
            while current+1 in nums:
                current += 1
                length += 1
            result = max(result,length)
    return result
````

````python
longest([0,1,2,3,5,7,6,100,8,9,8])
````

## WAP to  create frequency count for words in given sentance
    input - the quick brown fox jumps over the dog
    output - {'the' : 2, 'quick' : 1, ...}

````python
s = 'the quick brown fox jumps over the dog' 
s = s.split(" ")
count = 0
m = {}
for i in s:
        count = 0
        temp = i
        for j in s:
            if temp == j:
                count += 1
        m[i] = count
print(m)
````

````python
inp = 'the quick brown fox jumps over the dog'
freq = {i: inp.count(i) for i in inp.split()}
print(freq)
````

````python
inp = 'the quick brown fox jumps over the dog'
d = {}
l = inp.split()
for i in l:
    d[i] = d.get(i,0)+1
print(d)
````

### WAP to get single digits in numbers sorted backward and converted to english words
```
input - [1,3,4,5,11,12]
output - ['five','four','three','one']
